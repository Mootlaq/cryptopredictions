```python
import pandas as pd
import numpy as np
import requests
from pycoingecko import CoinGeckoAPI
import datetime
import cufflinks as cf
import matplotlib.pyplot as plt
import altair as alt
cg = CoinGeckoAPI()
import time
pd.set_option('display.max_rows', None)
```

```python
def get_history(coin):
    coin_history = cg.get_coin_market_chart_by_id(id=coin, vs_currency='usd', days='max')

    days = [i[0] for i in coin_history['prices']]
    days = pd.to_datetime(days, unit='ms')

    prices = [i[1] for i in coin_history['prices']]

    days = days[:-1]
    prices = prices[:-1]


    df = pd.DataFrame(data={'{} Price'.format(coin.capitalize()): prices}, index=days)
    
    return df

    
```

```python
def mc_projection(df, coin, projected_price): 
    btc_market_cap = final_df.at['btc', 'market_cap']
    eth_market_cap = final_df.at['eth', 'market_cap']
    old_market_cap = final_df.at[coin, 'market_cap']
    mc_at_projected = projected_price*final_df.at[coin, 'circulating_supply']
    
    coins = ['btc', 'eth', '{} projected marketcap'.format(coin), '{} actual marketcap'.format(coin)]
    marketcaps = [btc_market_cap, eth_market_cap, mc_at_projected, old_market_cap]
    images = [final_df.at['btc', 'image'], final_df.at['eth', 'image'], final_df.at[coin, 'image'], final_df.at[coin, 'image']]
    new_df = pd.DataFrame({'name': coins, 'market_cap': marketcaps, 'images':images})
    
    return eth_market_cap, mc_at_projected, new_df
    
```
